```python
import colorama
from wfc import parse_example_matrix, CompatibilityOracle, Model, render_colors

input_matrix =[
'LLLLLLLLLL',
'LLLLL:LLLL',
':L::::::LL',
'::::~~::::',
'~~~~~~~~~~',
'~~~~~~~~~~',
]

colors = {
    'L': colorama.Fore.GREEN,
    '~': colorama.Fore.BLUE,
    ':': colorama.Fore.YELLOW,
    'V': colorama.Fore.BLACK,
}

compatibilities, weights = parse_example_matrix(input_matrix)
compatibility_oracle = CompatibilityOracle(compatibilities)
model = Model((30, 75), weights, compatibility_oracle)
output = model.run()

render_colors(output, colors)
```
